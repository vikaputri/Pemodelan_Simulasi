# KEDATANGAN

```
ta = []
n = 10
for i in range(n):
  if i == 0:
    hasil = 0.5
  else:
    hasil += 0.5
  ta.append(hasil)
```

```
fa = [2, 4, 9, 22, 24, 16, 11, 6, 4, 2]
pa = []
pakum = []
bagi = sum(fa)
for i in range(len(fa)):
  hasil = fa[i]/bagi
  pa.append(hasil)
  result =sum(pa) 
  pakum.append(result)
```

```
import numpy as np
inta = []
for i in range(len(pakum)):
  if i==0:
    result = list((0.0, pakum[i]-0.00000000000000001))
  elif i > 0:
    result = list((pakum[i-1], pakum[i]-0.000000000000001))
  inta.append(result)
```

```
import pandas as pd
d = {'ta':ta,'pa':pa,'pakum':pakum, 'inta (awal, akhir)':inta}
df1 = pd.DataFrame(d)
df1
```

```
pelanggan = []
n = 31
for i in range(1, n+1):
  hasil = i
  pelanggan.append(hasil)
```

```
def random(a, m, c, x0, n):
	x = []
	x.append(x0)
	for i in range(n):
		rand = (a * x[i] + c) % m
		x.append(rand)
	return x
```

```
a = 7
m = 99
c = 6
x0 = 3
n = 30
xi = random(a, m, c, x0, n)
bil = [(i/m) for i in xi]
```

```
ta_hasil = []
for i in bil:
  if i < pakum[0]:
    result = ta[0]
  elif i > pakum[0] and i < pakum[1]:
    result = ta[1]
  elif i > pakum[1] and i < pakum[2]:
    result = ta[2]
  elif i > pakum[2] and i < pakum[3]:
    result = ta[3]
  elif i > pakum[3] and i < pakum[4]:
    result = ta[4]
  elif i > pakum[4] and i < pakum[5]:
    result = ta[5]
  elif i > pakum[5] and i < pakum[6]:
    result = ta[6]
  elif i > pakum[6] and i < pakum[7]:
    result = ta[7]
  elif i > pakum[7] and i < pakum[8]:
    result = ta[8]
  elif i > pakum[8] and i < pakum[9]:
    result = ta[9]
  ta_hasil.append(result)
```

```
wd = []
jam = 10
menit = 0
detik = 0
for i in ta_hasil:
  jam = jam
  a = i//1
  menit = menit + a
  b = (i% 1) * 60
  detik = detik + b
  if menit >= 60 :
    jam = jam + 1
    menit = menit - 60
  if detik >= 60 :
    menit = menit + 1
    detik = detik -60
  result_wd = "%d:%02d:%02d" % (jam, menit, detik) 
  wd.append(result_wd)
```

```
wp = ['10:01:00', '10:03:00', '10:07:30', '10:11:00', '10:11:30', '10:13:00', '10:15:30',
      '10:18:00', '10:20:00','10:23:00', '10:25:00', '10:27:30', '10:29:30', '10:31:30', 
      '10:33:30', '10:36:00', '10:39:30', '10:43:00', '10:45:30', '10:47:00', '10:50:00',
      '10:54:05', '10:57:05', '10:59:35', '11:02:35', '11:05:40', '11:07:40', '11:09:15',
      '11:13:15', '11:15:45', '11:16:45']
tmp = []
for i in range(len(wd)): 
    if wd[i] in wp: 
      p = 0
    else:
      for j in range(len(wp)):
        j = i
        a = wp[j]
        (h, m, s) = a.split(':')

        c = int(h) * 3600 + int(m) * 60 + int(s)
        b = wd[i]
        (h, m, s) = b.split(':')
        d = int(h) * 3600 + int(m) * 60 + int(s)
        p = c - d
    tmp.append(p)     
```

```
wd = []
jam = 10
menit = 0
detik = 0
for i in ta_hasil:
  jam = jam
  a = i//1
  menit = menit + a
  b = (i% 1) * 60
  detik = detik + b
  if menit >= 60 :
    jam = jam + 1
    menit = menit - 60
  if detik >= 60 :
    menit = menit + 1
    detik = detik -60
  result_wd = "%d:%02d:%02d" % (jam, menit, detik) 
  wd.append(result_wd)
```

```
PA = []
WT = []
for i in range(len(tmp)):
  if tmp[i] == 0 :
    hasil_pa = 0
    hasil_wt = 0
  else:
    if tmp[i] == tmp[i-1]:
      hasil_pa = 0
      hasil_wt = 0
    else:
      hasil_pa = 1
      hasil_wt = tmp[i]-tmp[i-1]
  PA.append(hasil_pa)
  WT.append(hasil_wt)
```

```
d = {'Pelanggan':pelanggan,'Bilangan Acak':bil,'ta':ta_hasil,'Waktu Datang':wd, "Waktu Pelayanan":wp, "Panjang Antrean":PA, "Waktu Tunggu":WT}
df4= pd.DataFrame(d)
df4
```

```
def random(a, m, c, x0, n):
	x = []
	x.append(x0)
	for i in range(n):
		rand = (a * x[i] + c) % m
		x.append(rand)
	return x
```

```
a = 2
m = 33
c = 8
x0 = 7
n = 30
xi = random(a, m, c, x0, n)
bilcak = [(i/m) for i in xi]
```

```
ts_hasil = []
for i in bilcak:
  if i < skum[0]:
    result1 = ts[0]
  elif i > skum[0] and i < skum[1]:
    result1 = ts[1]
  elif i > skum[1] and i < skum[2]:
    result1 = ts[2]
  elif i > skum[2] and i < skum[3]:
    result1 = ts[3]
  elif i > skum[3] and i < skum[4]:
    result1 = ts[4]
  elif i > skum[4] and i < skum[5]:
    result1 = ts[5]
  ts_hasil.append(result1)
```

```
import time 
wap = []
wap1 = []
jam1 = 0
for k in range(len(ts_hasil)):
  jam1 = jam1
  menit1 = ts_hasil[k]//1
  detik1 = (ts_hasil[k] % 1) * 60
  result2 = "%d:%02d:%02d" % (jam1, menit1, detik1)
  wapp = int(jam1) * 3600 + int(menit1) * 60 + int(detik1)
  for l in range(len(wp)):
    l = k 
    a1 = wp[l]
    (h, m, s) = a1.split(':')
    wdd = int(h) * 3600 + int(m) * 60 + int(s)
    jum = wdd + wapp
  h = time.strftime('%H:%M:%S', time.gmtime(jum))
  wap.append(h)
  wap1.append(result2)
```

```
WL = []
for n in range(len(ts_hasil)):
  if n == 0:
    hazil = ta_hasil[0]
  else:
    hazil = ta_hasil[n] - ts_hasil[n-1]
    if hazil < 0 :
      hazil = 0
  WL.append(hazil)
```

```
d = {'Bilangan Acak':bilcak,'ts':ts_hasil, 'Waktu Akhir Pelayanan':wap, 'WAKTU LUANG':WL}
df5 = pd.DataFrame(d)
df5
```
